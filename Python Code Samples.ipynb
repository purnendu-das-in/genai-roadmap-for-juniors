### 🔹 2.1 Variables & Data Types

A **variable** is just a labeled box that stores a value. Python figures out the type automatically.

| Type | Example | Use case |
|------|---------|----------|
| `str` | `"Hello"` | Text, prompts, responses |
| `int` | `42` | Token counts, indices |
| `float` | `0.7` | Temperature, probabilities |
| `bool` | `True` / `False` | Flags, conditions |
| `None` | `None` | Missing/empty values |


````python
# 🔹 Variables — labels for values
name = "Aarav"            # str: text in quotes
age = 25                  # int: whole number
temperature = 0.7         # float: decimal — used a LOT in LLMs!
is_beginner = True        # bool: True or False (capital T/F!)
favorite_model = "gpt-4o" # str

# Use f-strings (formatted strings) to mix variables into text — super clean
intro = f"Hi, I'm {name}, age {age}. My LLM temp is {temperature}."
print(intro)

# Check the type of any variable
print("Type of temperature:", type(temperature))
````

### 🔹 2.2 Lists — Ordered Collections

A **list** is an ordered, changeable collection. We use lists EVERYWHERE in GenAI: chat history, document chunks, embeddings, search results.


````python
# 🔹 Lists — ordered, changeable, allow duplicates
# Lists use square brackets [ ]
prompts = ["Summarize this", "Translate to French", "Write a tweet"]

# Indexing starts at 0
print("First prompt:", prompts[0])      # "Summarize this"
print("Last prompt:",  prompts[-1])     # negative indexes count from the end

# Add an item with .append()
prompts.append("Explain like I'm 5")
print("After append:", prompts)

# Length of the list
print("Total prompts:", len(prompts))

# Slicing: get items from index 1 to (but not including) index 3
print("Slice [1:3]:", prompts[1:3])
````

### 🔹 2.3 Dictionaries — Key/Value Pairs

A **dictionary** stores data as `key: value` pairs. This is the **#1 most important data structure for working with APIs and LLMs** — JSON responses come back as dictionaries!


````python
# 🔹 Dictionaries — key:value pairs (uses curly braces { })
# This is exactly what an LLM API response looks like
llm_response = {
    "model":  "gemini-1.5-flash",  # which model was used
    "tokens": 142,                 # how many tokens were consumed
    "text":   "Hello, world!",     # the actual generated content
    "finish_reason": "stop"        # why the model stopped
}

# Access a value by key
print("Model:", llm_response["model"])

# Safer access with .get() — returns None if the key doesn't exist
print("Cost:", llm_response.get("cost", "N/A"))   # 2nd arg = default

# Add or update a key
llm_response["temperature"] = 0.7
print("Updated dict:", llm_response)

# Loop over keys and values
for key, value in llm_response.items():
    print(f"  • {key} → {value}")
````

### 🔹 2.4 Loops — Repeating Actions

When you have multiple prompts, multiple files, or multiple chat turns, you'll use loops to process them all.


````python
# 🔹 The for-loop is by far the most common loop in Python

# Loop over a list
questions = ["What is AI?", "What is RAG?", "What are embeddings?"]
for q in questions:                # 'q' is the loop variable
    print("❓ Question:", q)

print("---")

# Loop with index using enumerate()
for i, q in enumerate(questions, start=1):   # start counting from 1
    print(f"Q{i}: {q}")

print("---")

# range(n) gives numbers 0, 1, 2, ..., n-1
for n in range(3):                  # 3 iterations
    print(f"Retry attempt #{n+1}")
````

### 🔹 2.5 Functions — Reusable Blocks of Code

Functions let you **bundle code** so you can reuse it. Every GenAI app you'll ever write is built from small functions.


````python
# 🔹 Defining a function with 'def'
def greet(name, language="English"):
    """Return a greeting in the given language. Default = English."""
    greetings = {
        "English":  f"Hello, {name}!",
        "Spanish":  f"¡Hola, {name}!",
        "Japanese": f"こんにちは, {name}さん!"
    }
    # .get() returns the English version if language not found
    return greetings.get(language, greetings["English"])

# Call the function with different arguments
print(greet("Aarav"))                      # uses default 'English'
print(greet("Maria",  "Spanish"))
print(greet("Yuki",   "Japanese"))

# Functions are how we'll wrap LLM calls — preview:
def ask_llm(prompt: str) -> str:
    """A placeholder for now — we'll fill this in section 3."""
    return f"(Pretend response to: '{prompt}')"

print(ask_llm("Tell me a joke"))
````

### 🔹 2.6 String Manipulation — Cleaning Text Like a Pro

LLMs are all about text. You'll constantly be cleaning, slicing, and reformatting strings.


````python
# 🔹 Common string operations you'll use daily
text = "  Generative AI is AMAZING!!  "

print("Original :", repr(text))            # repr() shows hidden chars
print("Stripped :", text.strip())          # remove whitespace
print("Lower    :", text.lower())          # all lowercase
print("Replaced :", text.replace("!", ""))  # remove exclamation marks
print("Split    :", text.strip().split())   # break into list of words
print("Length   :", len(text))             # character count

# Combining (joining) a list back into a string
words = ["AI", "is", "the", "future"]
sentence = " ".join(words)                  # join with spaces
print("Joined   :", sentence)

# f-string with formatting
score = 0.876543
print(f"Confidence: {score:.2%}")           # 87.65%
````

### 🔹 2.7 JSON Handling — The Language of APIs

**JSON** (JavaScript Object Notation) looks just like a Python dict. It's how APIs send and receive data.


````python
# 🔹 Working with JSON — convert between Python dict and JSON string
import json   # built-in module — no install needed

# A typical chat message format used by OpenAI/Gemini
message = {
    "role": "user",
    "content": "Hello AI",
    "metadata": {"timestamp": "2026-05-09", "lang": "en"}
}

# dict → JSON string (for sending to APIs)
json_string = json.dumps(message, indent=2)   # indent for pretty printing
print("JSON string:")
print(json_string)
print(type(json_string))   # <class 'str'>

print("---")

# JSON string → dict (for reading API responses)
parsed = json.loads(json_string)
print("Parsed back to dict:", parsed)
print("Just the role:",       parsed["role"])
````

### 🌐 What Is an API?

An **API** (Application Programming Interface) is a *contract* that lets one program talk to another.

Think of it like a **restaurant menu** 🍕:
- 📋 **You** = the client (your Python code)
- 🧑‍🍳 **Kitchen** = the server (e.g., OpenAI/Gemini)
- 📜 **Menu** = the API documentation
- 🛎️ **Waiter** = the API itself, taking your order and bringing back food

You don't need to know HOW the kitchen cooks — you just call the API and get a result.

```mermaid
sequenceDiagram
    participant You as Your Python Code
    participant API as LLM API
    participant Model as AI Model
    You->>API: POST request (prompt + key)
    API->>Model: process prompt
    Model->>API: generate response
    API->>You: JSON response
```


### 🌐 REST API Basics

Most APIs you'll use are **REST APIs**. The 4 core verbs:

| Verb | Meaning | Example |
|------|---------|---------|
| `GET` | Read data | Fetch a webpage |
| `POST` | Send data | Send a prompt to an LLM |
| `PUT` | Update | Edit your profile |
| `DELETE` | Remove | Delete a record |

For LLMs you'll mostly use **POST** — you're sending a prompt and getting back a response.


````python
# 🌐 Make a basic GET request to a free public API (no key needed)
import requests   # the most popular HTTP library in Python

# Free joke API — perfect for testing
url = "https://official-joke-api.appspot.com/random_joke"

response = requests.get(url)            # send GET request
print("Status code:", response.status_code)   # 200 = OK
print("Type      :", type(response))

# Parse the JSON body into a Python dict
joke = response.json()
print("\n😂 Random joke:")
print("Setup     :", joke["setup"])
print("Punchline :", joke["punchline"])
````

### 🤖 Calling an LLM (Gemini)

Now let's send a real prompt to a real AI model.


````python
# 🤖 Reusable function to call Gemini — we'll keep using this
import google.generativeai as genai
import os

# Configure once with our API key
genai.configure(api_key=os.environ['GEMINI_API_KEY'])

def ask_gemini(prompt: str, model_name: str = "gemini-1.5-flash",
               temperature: float = 0.7) -> str:
    """Send a prompt to Gemini and return the text reply.

    Args:
        prompt: the question or instruction for the model.
        model_name: which Gemini model to use.
        temperature: 0 = deterministic, 1 = creative (more on this below).
    Returns:
        The model's text response as a plain string.
    """
    model = genai.GenerativeModel(
        model_name,
        generation_config={"temperature": temperature}
    )
    response = model.generate_content(prompt)
    return response.text

# Try it out!
reply = ask_gemini("Explain APIs to a 10-year-old in 3 sentences.")
print(reply)
````

### 🌡️ Temperature — Controlling Creativity

**Temperature** is the most important LLM setting. It controls *how random* the output is.

| Temperature | Behavior | When to use |
|:---:|:---|:---|
| `0.0` | Deterministic, factual | Q&A, classification, code |
| `0.3` | Mostly factual | Summaries, structured data |
| `0.7` | Balanced ⚖️ (default) | Chatbots, general use |
| `1.0+` | Highly creative, unpredictable | Stories, brainstorming |


````python
# 🌡️ Compare temperatures with the same prompt
prompt = "Give me a creative product name for an AI study assistant app."

print("🔵 Temperature = 0.0 (boring & repeatable)")
print(ask_gemini(prompt, temperature=0.0))

print("\n🟠 Temperature = 1.2 (wild & creative)")
print(ask_gemini(prompt, temperature=1.2))
````

### 🎟️ Tokens — The Currency of LLMs

LLMs don't see words — they see **tokens** (chunks of characters).

- 1 token ≈ **4 characters** in English
- 1 token ≈ **0.75 words**
- "Hello world" ≈ 2 tokens
- A 500-word email ≈ ~650 tokens

You're billed **per token**, both for input (your prompt) AND output (the reply).

<div style="background:#fee2e2; border-left:5px solid #dc2626; padding:14px; border-radius:6px; color:#7f1d1d;">
<b>💸 Money tip:</b> Long system prompts + huge contexts = expensive. Always estimate tokens before sending big requests.
</div>


````python
# 🎟️ Count tokens for any text using tiktoken (OpenAI's tokenizer)
import tiktoken

# 'cl100k_base' is the encoder used by GPT-4o, GPT-3.5
encoder = tiktoken.get_encoding("cl100k_base")

samples = [
    "Hello",
    "Hello world",
    "Generative AI is changing software development forever.",
]

for text in samples:
    tokens = encoder.encode(text)         # text → list of token IDs
    print(f"{text!r:<60} → {len(tokens)} tokens")
````

### 💬 Mini Chatbot (Multi-Turn Memory)

Real chatbots remember the conversation. We do that by sending the WHOLE history every turn.


````python
# 💬 Tiny multi-turn chatbot using Gemini's chat session
chat = genai.GenerativeModel('gemini-1.5-flash').start_chat(history=[])

def chat_turn(user_msg: str) -> str:
    """Send one user turn, return assistant reply (history is auto-managed)."""
    response = chat.send_message(user_msg)
    return response.text

# Try a 3-turn conversation — the model remembers earlier turns
print("🧑 You:", "My name is Aarav and I love pizza 🍕")
print("🤖 AI :", chat_turn("My name is Aarav and I love pizza 🍕"))

print("\n🧑 You:", "What's my favorite food?")
print("🤖 AI :", chat_turn("What's my favorite food?"))

print("\n🧑 You:", "Suggest 3 pizza toppings I should try.")
print("🤖 AI :", chat_turn("Suggest 3 pizza toppings I should try."))
````

### 🎭 4.1 Role Prompting

Telling the LLM *who* it is dramatically improves quality. This is the single highest-ROI prompt technique.


````python
# 🎭 Same task, two prompts — see the difference
bad_prompt  = "Review this resume: 'Worked at company. Did stuff.'"

good_prompt = """You are a senior tech recruiter at FAANG with 15 years of experience.

Review this resume bullet and:
1. Score it from 1-10
2. List 2 specific weaknesses
3. Rewrite it using strong action verbs and metrics

Resume bullet: 'Worked at company. Did stuff.'"""

print("❌ BAD prompt response:")
print(ask_gemini(bad_prompt))
print("\n" + "="*60 + "\n")
print("✅ GOOD prompt response:")
print(ask_gemini(good_prompt))
````

### 🎯 4.2 Few-Shot Prompting

Give the model 2-3 **examples** of the input/output pattern you want — it'll mimic the style perfectly.


````python
# 🎯 Few-shot example: classify customer feedback as Positive / Negative / Neutral
few_shot_prompt = """Classify the sentiment as Positive, Negative, or Neutral.

Example 1:
Feedback: "Loved the new dashboard, super clean!"
Sentiment: Positive

Example 2:
Feedback: "App crashes every time I open it. Frustrating."
Sentiment: Negative

Example 3:
Feedback: "It works fine, nothing special."
Sentiment: Neutral

Now classify this:
Feedback: "The pricing page is confusing but the product is solid."
Sentiment:"""

print(ask_gemini(few_shot_prompt, temperature=0))
````

### 🧱 4.3 Structured Prompting

Use clear sections (`###`, `---`, headings) so the model knows exactly what each part does.

### 📄 5.2 PDF Summarizer App

Upload a PDF → get an instant AI summary. Useful for research papers, contracts, lecture notes.


````python
# 📄 PDF Summarizer with Gradio + pypdf
import gradio as gr
from pypdf import PdfReader

def extract_text_from_pdf(pdf_file) -> str:
    """Read every page and concatenate the text."""
    reader = PdfReader(pdf_file.name)         # Gradio gives us a file object
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text

def summarize_pdf(pdf_file, summary_length: str) -> str:
    """End-to-end: PDF file → extracted text → LLM summary."""
    if pdf_file is None:
        return "⚠️ Please upload a PDF first."

    # Extract — and truncate to ~12k chars to stay within free token limits
    text = extract_text_from_pdf(pdf_file)[:12000]

    # Build a clean structured prompt
    prompt = f"""You are an expert summarizer. Summarize the document below.

Length: {summary_length}
Output format:
- 🎯 Main idea (1 sentence)
- 📌 Key points (bullet list)
- 💡 Actionable takeaways (bullet list)

Document:
\"\"\"
{text}
\"\"\"
"""
    return ask_gemini(prompt, temperature=0.3)

# Build the UI
pdf_app = gr.Interface(
    fn=summarize_pdf,
    inputs=[
        gr.File(label="📄 Upload your PDF", file_types=[".pdf"]),
        gr.Radio(["Short", "Medium", "Detailed"], label="Summary length", value="Medium")
    ],
    outputs=gr.Markdown(label="📝 Summary"),
    title="📄 AI PDF Summarizer",
    description="Upload any PDF and get a clean AI-generated summary in seconds.",
    theme="soft",
)

pdf_app.launch(share=True, debug=False)
````

# 6️⃣ Section 6 — RAG Basics

<div style="background: linear-gradient(135deg, #14b8a6 0%, #0ea5e9 100%); padding:20px; border-radius:10px; color:white;">
<b>🎯 Goal:</b> Build a "Chat with PDF" app — your AI will answer questions based on YOUR documents, not its training data.
</div>


### 🤔 What Is RAG?

**RAG = Retrieval-Augmented Generation**

LLMs don't know your private documents. RAG fixes that by:
1. **Splitting** your docs into small chunks
2. **Converting** each chunk into a vector (embedding)
3. **Storing** vectors in a database
4. When the user asks a question → **retrieve** the most relevant chunks
5. **Stuff** those chunks into the prompt → LLM answers using YOUR data

```mermaid
flowchart LR
    A[📄 PDF] --> B[✂️ Chunk]
    B --> C[🔢 Embed]
    C --> D[(💾 Vector DB)]
    E[❓ Question] --> F[🔢 Embed]
    F --> G{🔍 Similarity Search}
    D --> G
    G --> H[📦 Top Chunks]
    H --> I[🤖 LLM]
    E --> I
    I --> J[💬 Answer]
```


### 🧩 Key Concepts

| Concept | What it is | Real-world analogy |
|---------|-----------|--------------------|
| **Chunking** | Split docs into small pieces | Cutting a book into paragraphs |
| **Embedding** | Turn text into a vector of numbers | A "fingerprint" of meaning |
| **Vector DB** | Stores embeddings for fast lookup | A library card catalog |
| **Semantic search** | Find chunks with similar meaning | Finding "car" when you search "vehicle" |


### 🛠️ Step 1 — Prepare a Sample Document

We'll create a sample document. (In a real app, replace this with a PDF you upload.)


````python
# 🛠️ Create a sample document about a fictional company
sample_doc = """
ACME GenAI Inc. — Internal Knowledge Base

ABOUT US
ACME GenAI Inc. was founded in 2024 by Aarav Sharma and Lina Park.
Our headquarters are in Bangalore, India, with satellite offices in
San Francisco and London. We currently employ 145 people.

PRODUCT LINE
Our flagship product is "SmartFan X200" — an AI-powered fan that adjusts
its speed based on room temperature and occupancy. It launched in March 2025
and has sold over 50,000 units in its first 6 months.

WARRANTY POLICY
All SmartFan X200 units come with a 2-year limited warranty covering
manufacturing defects. The warranty does NOT cover water damage, drops,
or unauthorized modifications. Customers can register their warranty
at acmegenai.example/warranty.

PRICING
- SmartFan X200 Basic: $129
- SmartFan X200 Pro:   $199 (includes voice control)
- Enterprise pack of 50 units: $5,500

SUPPORT HOURS
Customer support is available Monday-Friday, 9 AM to 8 PM IST.
Email: support@acmegenai.example | Phone: +91-80-1234-5678
"""

with open("knowledge.txt", "w", encoding="utf-8") as f:
    f.write(sample_doc)

print("✅ Sample document saved.")
print(f"📏 Length: {len(sample_doc)} characters")
````

### ✂️ Step 2 — Chunking the Document

We split the document into overlapping chunks so each chunk has a coherent meaning.


````python
# ✂️ Use LangChain's text splitter — handles overlap & sentence boundaries
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Read the document
with open("knowledge.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Configure the splitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,      # ~300 characters per chunk
    chunk_overlap=50,    # overlap so context isn't cut mid-sentence
    separators=["\n\n", "\n", ". ", " ", ""]   # try these in order
)

chunks = splitter.split_text(text)

print(f"📦 Created {len(chunks)} chunks\n")
for i, chunk in enumerate(chunks, 1):
    print(f"--- Chunk {i} ({len(chunk)} chars) ---")
    print(chunk[:120] + ("..." if len(chunk) > 120 else ""))
    print()
````

### 🔢 Step 3 — Create Embeddings

We turn each chunk into a vector using a free open-source model — no API needed.


````python
# 🔢 Generate embeddings using sentence-transformers (runs locally, free)
from sentence_transformers import SentenceTransformer

# 'all-MiniLM-L6-v2' = small, fast, surprisingly accurate
embedder = SentenceTransformer("all-MiniLM-L6-v2")

# Encode every chunk → a NumPy array of vectors
embeddings = embedder.encode(chunks)

print("📐 Embeddings shape:", embeddings.shape)
# (num_chunks, 384) — 384 numbers describe each chunk's "meaning"
print(f"   • {embeddings.shape[0]} chunks")
print(f"   • {embeddings.shape[1]} dimensions per vector")
print("\n🔍 First chunk vector (first 8 values):")
print(embeddings[0][:8])
````

### 💾 Step 4 — Store in a Vector Database (FAISS)

**FAISS** = Facebook AI Similarity Search. Lightning-fast, free, and runs locally.


````python
# 💾 Build a FAISS index for fast similarity search
import faiss
import numpy as np

# Convert to float32 — required by FAISS
vectors = np.array(embeddings).astype("float32")
dim = vectors.shape[1]    # 384

# Build a flat (brute-force) L2 index — perfect for small datasets
index = faiss.IndexFlatL2(dim)
index.add(vectors)         # add all chunk vectors

print(f"✅ FAISS index built with {index.ntotal} chunks.")
print(f"   Dimension: {dim}")
````

### 🔍 Step 5 — Semantic Search

Embed the user's question with the same model, then find the closest chunks.


````python
# 🔍 Helper to retrieve top-k most relevant chunks
def retrieve(query: str, k: int = 2) -> list[str]:
    """Return the k chunks most similar to the query."""
    # 1. Embed the question
    q_vec = embedder.encode([query]).astype("float32")
    # 2. Search the FAISS index
    distances, indices = index.search(q_vec, k)
    # 3. Map indices back to text
    return [chunks[i] for i in indices[0]]

# Test it
results = retrieve("How long is the warranty?", k=2)
for i, r in enumerate(results, 1):
    print(f"--- Match {i} ---")
    print(r)
    print()
````

### 🤖 Step 6 — The Full RAG Pipeline

Combine retrieval + LLM = answer grounded in your data.


````python
# 🤖 Full RAG: retrieve relevant chunks, then ask the LLM with that context
def rag_answer(question: str, k: int = 3) -> str:
    """Answer a question using only the retrieved chunks as context."""
    # 1. Retrieve the most relevant chunks
    relevant_chunks = retrieve(question, k=k)
    context = "\n\n".join(relevant_chunks)

    # 2. Build a prompt that ANCHORS the model to our data
    prompt = f"""You are a helpful assistant for ACME GenAI Inc.
Answer the question using ONLY the context below.
If the answer isn't in the context, say "I don't have that info in our knowledge base."

CONTEXT:
{context}

QUESTION: {question}

ANSWER:"""

    return ask_gemini(prompt, temperature=0.2)

# Try several questions
for q in [
    "Who founded ACME GenAI Inc.?",
    "How much does the SmartFan X200 Pro cost?",
    "Does the warranty cover water damage?",
    "What's the support phone number?",
    "Do you sell laptops?"   # not in the data — should refuse
]:
    print(f"❓ {q}")
    print(f"💬 {rag_answer(q)}")
    print("-" * 60)
````

### 📚 Step 7 — Bonus: Full "Chat with PDF" App

Let's combine everything into a real Gradio app where users upload a PDF and chat with it.


````python
# 📚 Full Chat-with-PDF app
import gradio as gr
from pypdf import PdfReader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer
import faiss, numpy as np

# Global state (held in module scope)
state = {"index": None, "chunks": None, "embedder": SentenceTransformer("all-MiniLM-L6-v2")}

def build_index(pdf_file):
    """Parse PDF → chunks → embeddings → FAISS index."""
    if pdf_file is None:
        return "⚠️ Please upload a PDF first."

    # Read PDF text
    reader = PdfReader(pdf_file.name)
    text = "\n".join((p.extract_text() or "") for p in reader.pages)

    # Chunk
    splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    chunks = splitter.split_text(text)

    # Embed
    vectors = state["embedder"].encode(chunks).astype("float32")

    # Index
    idx = faiss.IndexFlatL2(vectors.shape[1])
    idx.add(vectors)

    state["index"], state["chunks"] = idx, chunks
    return f"✅ Indexed {len(chunks)} chunks. Now ask anything about your PDF!"

def chat_with_pdf(question, history):
    """Gradio chat callback — uses the indexed PDF as context."""
    if state["index"] is None:
        return "⚠️ Please upload a PDF first."

    q_vec = state["embedder"].encode([question]).astype("float32")
    _, ids = state["index"].search(q_vec, 3)
    context = "\n\n".join(state["chunks"][i] for i in ids[0])

    prompt = f"""Answer the question using ONLY the context.
If unsure, say so.

CONTEXT:
{context}

QUESTION: {question}"""
    return ask_gemini(prompt, temperature=0.2)

# Build the UI with two tabs
with gr.Blocks(theme="soft", title="📚 Chat with PDF") as pdf_chat:
    gr.Markdown("# 📚 Chat with your PDF\nUpload a PDF, then ask questions about it.")
    with gr.Row():
        pdf_in = gr.File(label="📄 PDF", file_types=[".pdf"])
        upload_btn = gr.Button("🚀 Index PDF", variant="primary")
    status = gr.Markdown()
    upload_btn.click(build_index, inputs=pdf_in, outputs=status)

    gr.ChatInterface(fn=chat_with_pdf, examples=["Summarize this document", "What are the key takeaways?"])

pdf_chat.launch(share=True, debug=False)
````

### 🤔 What Is LangChain?

**LangChain** is a Python framework that gives you **lego blocks** for LLM apps. Instead of writing raw API calls every time, you compose pre-built components:

| Block | What it does |
|-------|--------------|
| 🤖 **LLM / ChatModel** | Wrapper around any LLM (Gemini, GPT, Claude...) |
| 📝 **Prompt Template** | Reusable prompts with variables |
| 🔗 **Chain** | Pipeline: input → step 1 → step 2 → output |
| 📚 **Retriever** | Fetch relevant docs (used in RAG) |
| 🛠️ **Tool** | A function the LLM can call (calculator, web search...) |
| 🧠 **Memory** | Track conversation history |
| 🎯 **Agent** | LLM that picks which tool to use |

```mermaid
flowchart LR
    A[📝 Prompt Template] --> B[🤖 LLM]
    B --> C[🧹 Output Parser]
    C --> D[✅ Final Result]
    style A fill:#fbbf24,stroke:#d97706,color:#000
    style B fill:#60a5fa,stroke:#2563eb,color:#000
```

### 📦 8.5 Structured Output (Pydantic)

Force the LLM to return a **typed object**, with validation. This is gold for production apps.


````python
# 📦 Define the EXACT shape you want back, with type hints
from pydantic import BaseModel, Field
from typing import List

class MovieReview(BaseModel):
    """Structured movie review extracted from text."""
    title:      str   = Field(description="The movie title")
    rating:     float = Field(description="Star rating from 0 to 10")
    pros:       List[str] = Field(description="Positive points")
    cons:       List[str] = Field(description="Negative points")
    recommend:  bool  = Field(description="Would the reviewer recommend it?")

# Tell the LLM to follow this schema — LangChain handles parsing for us
structured_llm = llm.with_structured_output(MovieReview)

review_text = """I just watched 'The Matrix' again — what a masterpiece!
The visuals still hold up after 25 years and the philosophy is mind-bending.
Some of the side actors feel a bit dated though. Overall I'd give it 9/10
and absolutely recommend it to anyone who hasn't seen it."""

# Get a real Python object back!
review = structured_llm.invoke(f"Extract a structured review from: {review_text}")

print("Type        :", type(review).__name__)
print("Title       :", review.title)
print("Rating      :", review.rating, "⭐")
print("Pros        :", review.pros)
print("Cons        :", review.cons)
print("Recommend?  :", review.recommend)
````

### 🧠 8.6 Memory — Multi-Turn Conversations

LangChain makes chat history trivial with `RunnableWithMessageHistory`.


````python
# 🧠 Add conversational memory to any chain
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

# Prompt with a slot for the running history
chat_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful coding tutor. Remember context across turns."),
    MessagesPlaceholder(variable_name="history"),   # ← injected automatically
    ("user", "{input}"),
])

chat_chain = chat_prompt | llm | StrOutputParser()

# Per-session memory store (in production, use Redis / Postgres)
session_store = {}
def get_history(session_id: str):
    if session_id not in session_store:
        session_store[session_id] = InMemoryChatMessageHistory()
    return session_store[session_id]

# Wrap the chain with auto-memory
chat_with_memory = RunnableWithMessageHistory(
    chat_chain,
    get_history,
    input_messages_key="input",
    history_messages_key="history",
)

# Try a 3-turn conversation — memory is automatic!
config = {"configurable": {"session_id": "user_42"}}

print("🧑:", "My name is Aarav, and I'm learning Python.")
print("🤖:", chat_with_memory.invoke({"input": "My name is Aarav, and I'm learning Python."}, config))

print("\n🧑:", "What are the 3 hardest topics for me to learn first?")
print("🤖:", chat_with_memory.invoke({"input": "What are the 3 hardest topics for me to learn first?"}, config))

print("\n🧑:", "Remind me — what's my name and what am I learning?")
print("🤖:", chat_with_memory.invoke({"input": "Remind me — what's my name and what am I learning?"}, config))
````

### 📚 8.7 Mini-RAG with LangChain (Way Cleaner Than Section 6!)

Section 6 wired RAG by hand. LangChain does it in **~20 lines** with built-in components.


````python
# 📚 Production-style RAG using LangChain components
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.runnables import RunnablePassthrough

# 1. Load the doc we created in Section 6
with open("knowledge.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()

# 2. Split into chunks (LangChain's splitter)
splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
docs = splitter.create_documents([raw_text])      # returns Document objects

# 3. Embed + store in FAISS — one line!
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(docs, embeddings)

# 4. Turn the store into a retriever (k=2 nearest chunks)
retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

# 5. RAG prompt
rag_prompt = ChatPromptTemplate.from_template("""
Answer the question using ONLY the context below. If unsure, say so.

CONTEXT:
{context}

QUESTION: {question}
""")

# 6. Build the full RAG chain with LCEL
def format_docs(docs):
    return "\n\n".join(d.page_content for d in docs)

rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | rag_prompt
    | llm
    | StrOutputParser()
)

# Try it!
print(rag_chain.invoke("Who founded ACME GenAI Inc. and when?"))
print("\n---\n")
print(rag_chain.invoke("What's the warranty period?"))
````

### 🛠️ 8.8 Tools — Letting the LLM Call Functions

A **Tool** is a Python function the LLM can decide to call. This is how "agents" work.


````python
# 🛠️ Define tools — just regular Python functions with type hints + docstrings
from langchain_core.tools import tool
from datetime import datetime

@tool
def get_current_time() -> str:
    """Return the current date and time as a human-readable string."""
    return datetime.now().strftime("%A, %d %B %Y, %I:%M %p")

@tool
def calculate(expression: str) -> str:
    """Safely evaluate a basic math expression like '2 + 2 * 3'."""
    # In production, never use eval() — use a real math parser
    try:
        # Allow only digits and operators for safety
        allowed = set("0123456789+-*/.() ")
        if not set(expression) <= allowed:
            return "Invalid characters in expression."
        return str(eval(expression))
    except Exception as e:
        return f"Error: {e}"

@tool
def word_count(text: str) -> int:
    """Count the number of words in a piece of text."""
    return len(text.split())

# Bundle the tools and bind them to the LLM
tools = [get_current_time, calculate, word_count]
llm_with_tools = llm.bind_tools(tools)

# Ask a question that requires a tool call
ai_msg = llm_with_tools.invoke("What's 137 multiplied by 24?")

# The LLM doesn't ANSWER directly — it returns a "tool_calls" plan
print("📞 Tool calls the LLM wants to make:")
for call in ai_msg.tool_calls:
    print(f"  → {call['name']}({call['args']})")
````

### 🤔 What Is LangGraph?

**LangGraph** is built ON TOP of LangChain. It models your AI app as a **graph** of nodes (steps) and edges (transitions). Think of it as a **flowchart you can run**.

| LangChain | LangGraph |
|-----------|-----------|
| Linear chains: A → B → C | Graphs: branches, loops, conditions |
| Stateless steps | Shared state across all nodes |
| Single-shot tasks | Multi-turn agents that "think" |
| Great for prompts/RAG | Great for agents, workflows, automation |

```mermaid
flowchart TD
    Start([🚀 START]) --> Agent{🤖 Agent thinks}
    Agent -->|Need a tool| Tool[🛠️ Run tool]
    Tool --> Agent
    Agent -->|Done| End([✅ END])
    style Start fill:#22c55e,color:#000
    style End fill:#a855f7,color:#fff
    style Agent fill:#fbbf24,color:#000
```

> 💡 **Why does this matter?** Real-world AI tasks aren't linear. Agents need to *think → act → observe → repeat* — exactly what graphs model perfectly.

### 🧱 9.2 The 3 Building Blocks

Every LangGraph app has 3 pieces:

1. **State** — a Python TypedDict that holds all data flowing through the graph
2. **Nodes** — functions that read/update the state
3. **Edges** — connections between nodes (can be conditional!)


````python
# 🧱 Build your first graph: a simple 2-node workflow
from typing import TypedDict
from langgraph.graph import StateGraph, START, END

# 1️⃣ Define the SHARED STATE — every node can read & update these fields
class GraphState(TypedDict):
    topic:   str   # input
    outline: str   # filled by node 1
    article: str   # filled by node 2

# 2️⃣ Define NODES — each is just a Python function
def create_outline(state: GraphState) -> dict:
    """Node 1: turn the topic into an outline."""
    response = llm.invoke(f"Create a 5-bullet outline for an article about: {state['topic']}")
    return {"outline": response.content}     # only return what changed

def write_article(state: GraphState) -> dict:
    """Node 2: turn the outline into a full article."""
    response = llm.invoke(
        f"Write a 150-word article using this outline:\n{state['outline']}"
    )
    return {"article": response.content}

# 3️⃣ Build the GRAPH
workflow = StateGraph(GraphState)
workflow.add_node("outliner", create_outline)
workflow.add_node("writer",   write_article)

# Define the EDGES (the flow)
workflow.add_edge(START, "outliner")        # entry point
workflow.add_edge("outliner", "writer")     # outliner → writer
workflow.add_edge("writer", END)            # writer → done

# Compile into a runnable
app = workflow.compile()

# Run it!
result = app.invoke({"topic": "Why every developer should learn vector databases"})

print("📋 OUTLINE:\n" + result["outline"])
print("\n" + "="*60)
print("\n📝 ARTICLE:\n" + result["article"])
````

### 🌿 9.3 Conditional Edges — Branching Logic

Real workflows have decisions. LangGraph supports conditional routing — the graph picks the next node based on state.


````python
# 🌿 Conditional routing: classify a customer message and dispatch to the right handler
from typing import TypedDict, Literal
from langgraph.graph import StateGraph, START, END

class SupportState(TypedDict):
    message:   str
    category:  str      # 'billing' | 'technical' | 'general'
    response:  str

def classify(state: SupportState) -> dict:
    """Classify the incoming message into one of 3 categories."""
    prompt = f"""Classify this customer message into ONE word:
billing | technical | general

Message: {state['message']}
Reply with the single category word, lowercase."""
    category = llm.invoke(prompt).content.strip().lower()
    # Default to 'general' for safety
    if category not in ("billing", "technical", "general"):
        category = "general"
    print(f"🏷️ Classified as: {category}")
    return {"category": category}

def handle_billing(state):
    response = llm.invoke(f"You are a billing specialist. Help with: {state['message']}").content
    return {"response": "[BILLING TEAM] " + response}

def handle_technical(state):
    response = llm.invoke(f"You are a tech support engineer. Help with: {state['message']}").content
    return {"response": "[TECH TEAM] " + response}

def handle_general(state):
    response = llm.invoke(f"You are a friendly support agent. Help with: {state['message']}").content
    return {"response": "[GENERAL TEAM] " + response}

# Router function — decides which node to go to next based on state
def route(state: SupportState) -> Literal["handle_billing", "handle_technical", "handle_general"]:
    return f"handle_{state['category']}"

# Build the graph
g = StateGraph(SupportState)
g.add_node("classifier",       classify)
g.add_node("handle_billing",   handle_billing)
g.add_node("handle_technical", handle_technical)
g.add_node("handle_general",   handle_general)

g.add_edge(START, "classifier")
# CONDITIONAL edge — uses the route() function to pick next node
g.add_conditional_edges("classifier", route)
# All handlers go to END
g.add_edge("handle_billing",   END)
g.add_edge("handle_technical", END)
g.add_edge("handle_general",   END)

support_app = g.compile()

# Try with 3 different messages
for msg in [
    "My credit card was charged twice last month, please help.",
    "The dashboard keeps crashing when I export to CSV.",
    "Just wanted to say thanks — the team has been super helpful!"
]:
    print(f"\n📩 Customer: {msg}")
    out = support_app.invoke({"message": msg, "category": "", "response": ""})
    print(f"💬 Reply: {out['response'][:200]}...")
    print("-" * 60)
````

### 🤖 9.4 ReAct Agent — Think → Act → Observe Loop

The most powerful LangGraph pattern: an **agent** that decides which tools to call, in what order, until it has enough info to answer.


````python
# 🤖 Pre-built ReAct agent with tools (uses the tools we made in Section 8)
from langgraph.prebuilt import create_react_agent

# Reuse the tools defined earlier: get_current_time, calculate, word_count
agent = create_react_agent(llm, tools=tools)

# Ask a question that requires MULTIPLE tool calls
question = "What's the current time, and also what is 12 * 47 - 89?"

# Stream the agent's thinking
print(f"❓ Question: {question}\n")
result = agent.invoke({"messages": [("user", question)]})

# Each message in the trace shows the agent's reasoning
for msg in result["messages"]:
    role = msg.__class__.__name__.replace("Message", "").upper()
    if hasattr(msg, "tool_calls") and msg.tool_calls:
        for tc in msg.tool_calls:
            print(f"🛠️ [TOOL CALL] {tc['name']}({tc['args']})")
    elif hasattr(msg, "content") and msg.content:
        preview = str(msg.content)[:200]
        print(f"📨 [{role}] {preview}")
    print()
````

### 💾 9.5 Persistent Memory with Checkpoints

LangGraph can **automatically save state** between turns — perfect for long-running chatbots.


````python
# 💾 Add persistent memory using a checkpointer
from langgraph.checkpoint.memory import MemorySaver

# In-memory checkpointer (use SqliteSaver / PostgresSaver in production)
memory = MemorySaver()

# Re-create the agent WITH memory enabled
agent_with_memory = create_react_agent(llm, tools=tools, checkpointer=memory)

# A 'thread_id' identifies one conversation — each user gets their own
config = {"configurable": {"thread_id": "user_99"}}

# Turn 1
r1 = agent_with_memory.invoke(
    {"messages": [("user", "My favorite number is 23. Remember it.")]},
    config
)
print("🤖 Turn 1:", r1["messages"][-1].content[:150])

# Turn 2 — different invoke, but the agent REMEMBERS!
r2 = agent_with_memory.invoke(
    {"messages": [("user", "What's my favorite number squared?")]},
    config
)
print("\n🤖 Turn 2:", r2["messages"][-1].content[:150])
````

````python
# 🏗️ Mini Research Agent — uses the ACME knowledge base + calculator
from langchain_core.tools import tool

# Tool 1: search the ACME knowledge base (uses our LangChain RAG retriever from 8.7)
@tool
def search_acme_kb(query: str) -> str:
    """Search the ACME GenAI Inc. internal knowledge base. Use for any
    question about ACME's products, pricing, warranty, or company info."""
    docs = retriever.invoke(query)
    return "\n\n".join(d.page_content for d in docs)

# Tool 2: reuse calculator from Section 8.8

# Build the research agent
research_tools = [search_acme_kb, calculate]
researcher = create_react_agent(llm, tools=research_tools, checkpointer=MemorySaver())

# Ask a multi-step question
question = """If a company wants to buy 3 enterprise packs of SmartFan X200,
how much will they pay total? Also tell me how long the warranty lasts."""

print(f"❓ {question}\n")

result = researcher.invoke(
    {"messages": [("user", question)]},
    {"configurable": {"thread_id": "demo"}}
)

# Show the agent's reasoning steps
for msg in result["messages"]:
    if hasattr(msg, "tool_calls") and msg.tool_calls:
        for tc in msg.tool_calls:
            print(f"🛠️ {tc['name']}({tc['args']})")
    elif hasattr(msg, "content") and msg.content and msg.__class__.__name__ == "AIMessage":
        print(f"\n💬 Agent reply:\n{msg.content}")
    elif msg.__class__.__name__ == "ToolMessage":
        print(f"📦 Tool returned: {str(msg.content)[:120]}...")
    print()
````

## 🎤 Interview Tips (GenAI-Specific Questions)

<details>
<summary>📌 <b>Q: What's the difference between embeddings and tokens?</b></summary>

Tokens are units of text the model reads. Embeddings are dense vectors representing the *meaning* of text. You tokenize for the LLM; you embed for similarity search.

</details>

<details>
<summary>📌 <b>Q: Why use RAG instead of fine-tuning?</b></summary>

RAG is faster, cheaper, easier to update (just re-embed new docs), and shows citations. Fine-tuning teaches *style/format*, not new facts.

</details>

<details>
<summary>📌 <b>Q: How do you reduce LLM hallucinations?</b></summary>

Lower temperature, ground answers with RAG, ask the model to cite sources, add "Say 'I don't know' if unsure", and validate output with structured schemas.

</details>

<details>
<summary>📌 <b>Q: What's the difference between system, user, and assistant messages?</b></summary>

System sets the persona/rules. User is the actual question. Assistant is the model's reply. Multi-turn = a list of these in order.

</details>

<details>
<summary>📌 <b>Q: How would you reduce API costs?</b></summary>

Cache repeated prompts, use smaller models for simple tasks, truncate context, batch requests, set max_tokens, and use streaming so users can stop early.

</details>
